```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

# Running time

```python
#Without the assumption
time1 = np.array([[0.05397489, 5.16724041, 3.30919597, 0.49021058],
 [0.11564565, 5.82778811, 1.35159357, 0.5671958 ],
 [0.0511586,  5.78052719, 1.32094076, 0.55225761],
 [0.1446218,  5.46503675, 1.00803263, 0.56611912]])
#Under equal covariance matrices
time2 = np.array([[ 0.092, 11.113,  4.379,  0.504],
 [ 0.071, 12.419,  1.807,  0.854],
 [ 0.092, 12.579,  1.328,  0.948],
 [ 0.069, 11.187,  1.372,  1.048]])


missing_rate = ['20%','40%','60%','80%']
plt.figure(figsize=(12, 4))
plt.subplot(121)
plt.plot(missing_rate, time1[:,0],marker = 'o', markersize = 5,
            linestyle = '-.', linewidth = 2, color = "#c1272d", label = 'DPER')
plt.plot(missing_rate, time1[:,1],marker = 'o', markersize = 5,
            linestyle = '-.', linewidth = 2, color = "#0000a7", label = 'MICE')
plt.plot(missing_rate, time1[:,2],marker = 'o', markersize = 5,
            linestyle = '-.', linewidth = 2, color = "green", label = 'MissForest')
plt.plot(missing_rate, time1[:,3],marker = 'o', markersize = 5,
            linestyle = '-.', linewidth = 2, label = 'Soft-Impute')
plt.ylabel('Thời gian chạy (giây)',fontsize = 11)
plt.xlabel('Tỉ lệ khuyết',fontsize = 11)
plt.title("Ma trận hiệp phương sai khác nhau")
plt.legend()

plt.subplot(122)
plt.plot(missing_rate, time2[:,0],marker = 'o', markersize = 5,
            linestyle = '-.', linewidth = 2, color = "#c1272d", label = 'DPER')
plt.plot(missing_rate, time2[:,1],marker = 'o', markersize = 5,
            linestyle = '-.', linewidth = 2, color = "#0000a7", label = 'MICE')
plt.plot(missing_rate, time2[:,2],marker = 'o', markersize = 5,
            linestyle = '-.', linewidth = 2, color = "green", label = 'MissForest')
plt.plot(missing_rate, time2[:,3],marker = 'o', markersize = 5,
            linestyle = '-.', linewidth = 2, label = 'Soft-Impute')
plt.ylabel('Thời gian chạy (giây)',fontsize = 11)
plt.xlabel('Tỉ lệ khuyết',fontsize = 11)
plt.title("Ma trận hiệp phương sai bằng nhau")
plt.legend()

plt.show()
```
